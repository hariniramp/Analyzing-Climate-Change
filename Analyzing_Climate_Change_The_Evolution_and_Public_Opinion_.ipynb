## Question 1: Does a state's political affiliation impact the way they answer certain questions about climate change?

**Hypothesis: Republican states are likely to have greater percentages opposing ideals of climate change and policies to mitigate global warming, than Democratic states.**

**Methodology**

First, let's look at the data with county-level granularity. We have YPCC's climate change survey data ('survey_data') which contains survey participants' views on climate change aggregated at the county level. We also have MIT's Election Lab data, at a county level('county_data').

A new column called 'rep_percent' is being added to county_data, representing the "republican percentage" or the percentage strength of Republican voters in that county. This has been done by dividing the number of Republican votes by total number of votes, for a year, per county. This is to observe the effect of a county's Republican voter strength on that county's climate change belief systems.

Then, the county_data is filtered for the two main political parties: Democrat and Republican.

```
# Initializing the 'rep_percent' column with NaNs
county_data['rep_percent'] = np.nan
# For each row, calculating the 'candidatevotes' as a percentage of 'totalvotes'
county_data['rep_percent'] = np.divide(np.array(county_data['candidatevotes']), np.array(county_data['totalvotes']) )
# Filtering out parties to represent only 'democrat' or 'republican' rows
county_data = county_data[ (county_data['party']=='democrat') | (county_data['party']=='republican') ].reset_index()
county_data.head(5)
```

The cell below creates a mini dataframe to map US states to their abbreviations. For instance: California maps to 'CA'.

```
state_mappings = pd.DataFrame(county_data, columns = ['state','state_po']).drop_duplicates()
state_mappings.head(2)
```

Currently, the dataframe column 'rep_percent' contains the rowwise percentage of candidate votes as a share of total votes, per county per year. When the party is 'democrat', it represents the democratic votes percentage and when the party is 'republican', it represents the republican votes percentage. 

The cell below transforms the column 'rep_percent' to contain only the rowwise percentage of Republican candidate votes as a share of total votes, per county per year.

```
# Obtaining rowwise percentage of republican party votes
republican_percent = list(county_data[county_data['party']=='republican']['rep_percent'])
republican_percent1 = []
# Transforming 'rep_percent' column with republican percentage
for i in republican_percent:
  republican_percent1.extend([i, i])
county_data['rep_percent'] = republican_percent1
county_data.head(4)
```

The cell below determines the county winners for each year (either Republican or Democrat) by grouping by 'county' and 'year', and finding the maximum by aggregating on 'candidatevotes'. The resulting row represents the winner for that county and year.

```
# Grouping by year and county, aggregating by maximum candidate votes
county_winners = county_data.groupby(['year','county'], sort = False)['candidatevotes'].max().reset_index()
# Merging with original county data to re-obtain columns
county_winners = pd.merge(county_winners, county_data,  how='left', left_on=['year','county','candidatevotes'], right_on = ['year','county','candidatevotes'])
county_winners.head()
```

The cell below finds the average Republican vote strength per county by grouping by county and averaging the 'rep_percent' variable across the years 2000-2016. This might be a changing number since Republican vote strength will likely change over the years. However, my belief is that this method will result in a relatively fair reporting of a county's average Republican vote strength.

```
# Grouping by county and averaging 'rep_percent' over all years (2000-2016)
county_political = county_winners.groupby(['county'], sort = False)['rep_percent'].mean().reset_index()
county_political.head(3)
```

The cell below declares county winners overall by aggregating by county and finding the mode of the 'party' column. So, over the years 2000-2016, the party that a particular county has voted for the most times, will be deemed as that county's political affiliation. There is a slight risk of this being a short-sighted view of a county's political preference, and potentially assessing 'swing' counties on their most recent political choices. However, since this is the available data I have to work with, I hope this assumption holds strong.

```
# Grouping and declaring party winners by county 
county_winners_overall = county_winners.groupby(['county'], sort = False)['party'].agg(lambda x: scipy.stats.mode(x)[0][0]).reset_index()
county_winners_overall.head()
```

In addition to the visualization above, we can perform statistical tests to test the hypothesis.This will give us statistical evidence for any conclusions we can draw. Here, the following are the null and alternate hypotheses.

**Null Hypothesis (H_0):** Both Republican and Democratic-leaning counties have the same average percentage of people opposing ideals of climate change and policies to mitigate global warming. 

**Alternative Hypothesis (H_a):** Republican counties are likely to have greater percentages opposing ideals of climate change and policies to mitigate global warming, than Democratic counties.

To test our our hypothesis, we will need to select an appropriate statistical test. I have chosen the two-sample z-test since we will be comparing two samples (anti-climate change population percentages for Democratic-leaning, Republican-leaning counties). I selected a large sample z-test since the sample sizes in our dataset (democrats: 448 and republican: 1406) are relatively quite large.

**Considerations:**
* The sample sizes are pretty imbalanced (448 vs 1406). The z-test results work best when the sample sizes are roughly equal. I will sample 400 records from both the samples randomly. However, it must be noted that the initial sets have pretty vastly different sizes.
* The sample points are mostly independent from each other. It might be noteworthy to consider that geographically close counties might influence one another. However, generally, it is okay to assume that these data points are independent.

```
# x1 is a county's Republican population, x2 is the county's Democratic population
x1 = list(survey_joined[survey_joined['party']=='republican']['teachGWOppose'])
x2 = list(survey_joined[survey_joined['party']=='democrat']['teachGWOppose'])
# Performing the z-test and obtaining p-value
ztest, pval = stests.ztest(random.sample(x1, 400), x2=random.sample(x2,400), value=0,alternative='two-sided')
print(float(pval))
if pval<0.05:
    print("reject null hypothesis H0")
else:
    print("fail to reject null hypothesis H0")
```

**Discussion:**

Setting a confidence level of 95%, we need to observe a p-value below 0.05 in order to determine that our null hypothesis can be rejected. As we can observe, upon conducting the two-sample z-test, we found that the p-value was extremely low at 6.380372397209615e-37. This presents us with very strong evidence to reject the null hypothesis. Thus, we can say with 95% confidence that Republican states are likely to have greater percentages opposing ideals of climate change and policies to mitigate global warming, than Democratic states.

**Observations:**

* An upward trend is visible but it is not clear if the data is an increasing function of time. The mean seems to be increasing with time but we can make sure with another test of trend stationarity called the Dicky-Fuller test. The cell below will perform this test. The code for the test and plots for stationarity have been taken from this [Kaggle notebook](https://www.kaggle.com/sumi25/understand-arima-and-tune-p-d-q).

```
# Function definition for testing time series stationarity
def test_stationarity(timeseries, window = 12, cutoff = 0.01):

    #Determing rolling statistics
    rolmean = timeseries.rolling(window).mean()
    rolstd = timeseries.rolling(window).std()

    #Plot rolling statistics:
    fig = plt.figure(figsize=(12, 8))
    orig = plt.plot(timeseries, color='blue',label='Original')
    mean = plt.plot(rolmean, color='red', label='Rolling Mean')
    std = plt.plot(rolstd, color='black', label = 'Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show()
    plt.savefig('Visualizations/stationarity_rq4.png')
    
    #Perform Dickey-Fuller test:
    print('Results of Dickey-Fuller Test:')
    dftest = adfuller(timeseries, autolag='AIC', maxlag = 20 )
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    pvalue = dftest[1]
    if pvalue < cutoff:
        print('p-value = %.4f. The series is likely stationary.' % pvalue)
    else:
        print('p-value = %.4f. The series is likely non-stationary.' % pvalue)
    
    print(dfoutput)
```

The cell below invokes the function defined above to test stationarity.

```
test_stationarity(train_temp['LandAverageTemperature'])
```

As we can see, the results of the Dickey-Fuller test indicate that the time series is likely stationary. This allows us to proceed with an ARIMA model without performing any transformation on the values of the time series.

**ARIMA Model:**

The AR (autoregressive) portion of the ARIMA model correlates with the partial autocorrelation lags. We can see that up until 1 lag, there is significant peak. We can also observe a milder peak in the 13th lag. This suggests an autoregressive process of order 12. Let us try AR(1) value since the last significant lag was at k=1.

The I (integrated) can be already set to 0, as it is an indicator of stationarity and we observed no stationarity by evidence of the Dicky-Fuller test.

The MA (moving average) portion: This model is one where the forecast depends on the residuals of the lagged forecast.

The following cell sets date as the global temperature dataframe's index and splits the whole dataframe into training and testing sets.

```
# Set global temperature index as 'date'
global_temp = global_temp.set_index('dt')
# Split into train and test
train = global_temp.loc['1760-01-01':'2010-01-01']
test = global_temp.loc['2010-01-01':]
```

The following cell might take a while to run. It trains the auto ARIMA model for the Land Average Temperature time series. It does so by training multiple ARIMA models with different configurations of (p,d,q) values. Currently, I have assigned the maximum p and q values to be 1, since we deemed the lag in the AR portion of the model to be 1 using the PACF graph. The q values of 0 and 1 are tested as well. I have restricted these values to 0,1 because of two reasons: i. they seem to be fair values of p,q using the ACF and PACF plots, ii. extending max values to 2 onwards results in an extremely large number of permutations which causes all heap memory to be used and the RAM to crash. Thus,a limited range of p,q values is selected.

The auto ARIMA model runs ARIMA for multiple permutations and selects the permutation that results in the lowest AIC (Akaike Information Criterion) and BIC (Bayesian Information Criterion) scores. 

```
# Training the auto arima model
model = auto_arima(train['LandAverageTemperature'], start_p=1, start_q=1,
                           max_p=1, max_q=1, m=12, seasonal=True,trace=True)
```

The following cell fits the model on the training data time series, and predicts a temperature forecast for 72 months or 6 years. 

```
# Fitting the arima model to train data
model.fit(train['LandAverageTemperature'])
# Predicting temperature for the next 6 years (72 periods)
future_forecast = model.predict(n_periods=72)
future_forecast = shift(future_forecast, 1, cval=np.NaN)
```

## Question 5: Do states that gain more energy burning fossil fuels have higher surface temperatures?

This question is asked in order to try testing the hypothesis that burning more fossil fuels increases surface temperatures. We must take into account some considerations. 

**Hypothesis: States that use less renewable energy face increased surface temperatures.**

* The rate of global warming has been increasing over the past 200 years. However, in terms of degrees Celsius, the temperature increase could be in pretty small increments. Thus, there might not be a huge visible correlation in the graphs between the amount of fossil fuel generation and surface temperature.

* It must be noted that some states might generally have higher surface temperatures than others by default. For instance, Alaska has a significantly lower temperature than all other states. Thus, even if it were to produce a lot of energy through fossil fuels, it would take a significant number of years before the average temperature became higher than that of other states, and showed the correlation we might expect on the graph.

So, we must note these factors in this particular study and consider that there may be many outliers and we might not see the correlation we expect.
